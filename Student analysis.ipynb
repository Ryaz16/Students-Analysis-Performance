# Now we check the confidenec Interval of scores of all

```python
import scipy.stats as stats
import math

# random values
np.random.seed(10)

# lets take sample size
sample_size  =1000

sample = np.random.choice(a=data['math score'],size=1000)
sample_mean =sample.mean()


# get z-crirical value
z_critical  =stats.norm.ppf(q =0.95)
print("z_critical",z_critical)

# get population standard deviation
pop_std_dev = data['math score'].std()

# cleck margin of error
margin_of_error =z_critical * (pop_std_dev/math.sqrt(sample_size))

# NOW def of confidence intervals

confidence_intervals = (sample_mean - margin_of_error,sample_mean + margin_of_error)

# lets print the resut
print("confidence_intervals:", end=" ")
print(confidence_intervals)
print("True mean::{}".format(data['math score'].mean()))
```

```python
# Assigning grades to the grade acording to the following criteria

# 0-40 grade E
# 41-60 grade D
# 61-7   grade c
# 71-80 grade B
# 81-90 grade A
# 91-100 grade grade O

def getgarde(percentage ,Status):
    if Status == "Fail":
        return "E"
    if (percentage >= 90):
        return 'o'
    if (percentage >= 80):
        return "A"
    if (percentage >= 70):
        return "B"
    if (percentage >= 60):
        return "C"
    if (percentage >= 40):
        return "D"
    
    else :
        return "E"

data['grade']=data.apply(lambda x:getgarde(x['percentage'],x['Status']),axis=1)        
data['grade'].value_counts()
```

# Model Training

```python
# Indepandent and depandent variable
```

```python
# Independent variables (features)
independent_vars = ['gender', 'race/ethnicity', 'parental level of education', 'lunch',
                    'test preparation course', 'math score', 'reading score',
                    'writing score', 'Pass_reading', 'Pass_writing', 'Total_score',
                    'percentage', 'Status', 'pass_maths']

# Dependent variable
dependent_var = 'grade'

# Create X (independent variables) and y (dependent variable)
X = data[independent_vars]
y = data[dependent_var]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
X_train
```

```python
X_train.shape,y_train.shape
```

```python
X_test.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler

mm = MinMaxScaler()

## feedind the indepandent variable into the scalar

X_train= mm.fit_transform(X_train)
X_test= mm.transform(X_test)
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.fit(X_train,y_train)
 
y_pred = model.predict(X_test)

```

```python
print('training accuracy',model.score(X_train,y_train))
print('test accuracy',model.score(X_test,y_test))
```

```python
# Confusion matrix

from sklearn.metrics import confusion_matrix

cm =confusion_matrix(y_test,y_pred)
print(cm)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_pred)
print(cr)
```
